```python
import matplotlib.pyplot as plt
import pandas as pd

# Load your dataset
df_nailpolish = pd.read_csv('/Users/huma/Downloads/nailpolish (1).csv')

# Round the prices to the nearest whole number
df_nailpolish['rounded_price'] = df_nailpolish['price'].round()

# Create a histogram for the rounded price distribution of nail polish
plt.figure(figsize=(10, 6))
df_nailpolish['rounded_price'].hist(bins=range(int(df_nailpolish['rounded_price'].min()), 
                                               int(df_nailpolish['rounded_price'].max()) + 1), 
                                    color='purple')

# Set the title and labels
plt.title('Price Distribution for Nail Polish')
plt.xlabel('Price (Rounded to Nearest Whole Number)')
plt.ylabel('Frequency')

# Optionally, disable the grid, if you prefer
plt.grid(False)

# Show the plot
plt.show()
```

```python
from collections import Counter

# Flatten the list of tags and count occurrences
all_tags = [tag for sublist in df_nailpolish['tag_list'] for tag in sublist]
tag_counts = Counter(all_tags)

# Convert the counter to a DataFrame for easier visualization
tag_counts_df = pd.DataFrame(tag_counts.items(), columns=['Tag', 'Count']).sort_values(by='Count', ascending=False)

# Create a bar plot for the most common tags
plt.figure(figsize=(12, 8))
plt.bar(tag_counts_df['Tag'], tag_counts_df['Count'], color='teal')
plt.title('Most Common Tags')
plt.xlabel('Tags')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd

# Load the CSV file to inspect the data
data = pd.read_csv("foundation.csv")
data.head(), data.columns
```

```python
# Brand comparison: Count the number of products per brand
brand_counts = data['brand'].value_counts().reset_index()
brand_counts.columns = ['Brand', 'Number of Products']
brand_counts.head()
```

```python
# Price comparison with each product type: Calculate the average price for each product type
price_by_type = data.groupby('product_type')['price'].mean().reset_index()
price_by_type.columns = ['Product Type', 'Average Price']
price_by_type.sort_values('Average Price', ascending=False)
```

```python
# Product category price comparison: Calculate the average price for each category
price_by_category = data.groupby('category')['price'].mean().reset_index()
price_by_category.columns = ['Category', 'Average Price']
price_by_category.sort_values('Average Price', ascending=False)
```

```python
from ast import literal_eval

# Correctly parse the 'tag_list' column, which is in string representation of lists
data['parsed_tags'] = data['tag_list'].apply(literal_eval)

# Expand the list of tags into separate rows for each tag
all_tags = data['parsed_tags'].explode()

# Count the occurrence of each tag
tag_counts = all_tags.value_counts().reset_index()
tag_counts.columns = ['Tag', 'Count']
tag_counts.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
lipstick_data = pd.read_csv('/Users/huma/Downloads/lipstick (2).csv')

# Assuming your dataset has a 'category' column that includes these specific categories:
# 'lip_gloss', 'lip_stain', 'lipstick', and 'liquid' - adjust if your data uses different categories
categories = ['lip_gloss', 'lip_stain', 'lipstick', 'liquid']
# Filter the dataset to include only the relevant categories
lipstick_data_filtered = lipstick_data[lipstick_data['category'].isin(categories)]

# Calculate the average price for each of these categories
average_price_by_category = lipstick_data_filtered.groupby('category')['price'].mean().reset_index()

# Bar chart for Average Price by Lipstick Category
plt.figure(figsize=(10, 6))
plt.bar(average_price_by_category['category'], average_price_by_category['price'], color='salmon')
plt.xlabel('Lipstick Category')
plt.ylabel('Average Price ($)')
plt.title('Average Price by Lipstick Category')
plt.show()
```
